````python
headers = ['日期','幣別','現金匯率本行買入','現金匯率本行賣出','即期匯率本行買入','即期匯率本行賣出']

def get_exchange(url):
    res = requests.get(url)
    soup = BeautifulSoup(res.text,'lxml')
    df = pd.read_html(soup.select('table')[0].prettify())[0]
    df = df.drop(df.columns[6],axis=1)
    df.columns = headers
    return df
````

````python
from datetime import datetime,timedelta
````

````python
df = pd.DataFrame([])
for i in range(0,6):    
    dt = ( datetime.now() - timedelta(days= 30 * i) ).strftime('%Y-%m')
    url = 'https://rate.bot.com.tw/xrt/quote/{}/USD'.format(dt)
    df_ = get_exchange(url)
    df = pd.concat([df,df_])
````

````python
df = df.reset_index(drop=True)
````

````python
from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base
````

````python
from sqlalchemy import Date,Float

class Pcc(Base):
    __tablename__ = 'pcc'
    id = Column(Integer,primary_key=True,autoincrement=True)
    name = Column(String(100))
    casename = Column(String(100))
    method = Column(String(100))
    public_date = Column(Date)
    end_date = Column(Date)
    amount = Column(Float)
````

````python
Base.metadata.create_all(engine)
````

````python
engine.execute('SHOW TABLES;').fetchall()
````

````python
tmp_data = df.to_dict(orient='record')[0:5]
````

````python
tmp_data[0]
````

````python
import re
def datetime_parser(x):
    m =re.match('(\d{1,3})\/(\d{1,2})\/(\d{1,2})',x)
    dt = '%s/%s/%s'%(int(m.group(1))+1911,m.group(2),m.group(3))
    return datetime.strptime(dt,'%Y/%m/%d')
#     return datetime.strptime(x,'%Y/%m/%d')
````

````python
dt = '107/08/06'
datetime_parser(dt)
````

````python
def transform_data(row):
    name = row['機關  名稱']
    casename = row['標案案號  標案名稱']
    method = row['招標  方式']
    public_date = datetime_parser(row['公告  日期'])
    end_date = datetime_parser(row['截止  投標'])
    amount = row['預算  金額']
    return Pcc(name=name,casename=casename,method=method,public_date=public_date,end_date=end_date,amount=amount)
````

````python
from sqlalchemy.orm import sessionmaker
Session = sessionmaker(bind=engine)
session = Session()
````

````python
for row in tmp_data:
    doc =transform_data(row)
    session.add(doc)
    session.commit()
````

````python
for i in session.query(Pcc):
    print(i.casename,i.name)
````
